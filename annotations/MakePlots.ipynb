```python
import pandas as pd

df = pd.read_csv("../all_pairs - old.tsv", sep='\t')
print (len(df))
```

```python
label = 'Contains a 2'
#change_outcomes
```

```python
df["label"] = df.apply(lambda row: str(row[label]), axis=1)
df.head()
mean = df.groupby(['label'])['difflib standard change_ratio'].mean()
std = df.groupby(['label'])['difflib standard change_ratio'].std()

mean.plot(kind='bar', yerr=std)
```

```python
column_name = "new_change_ratio (re-arrangements at 0.5)"
mean = df.groupby(['label'])[column_name].mean()
std = df.groupby(['label'])[column_name].std()

mean.plot(kind='bar', yerr=std)
```

```python
column_name = "Word_change_ratio"
mean = df.groupby(['label'])[column_name].mean()
std = df.groupby(['label'])[column_name].std()

mean.plot(kind='bar', yerr=std)
```

```python
column_name = "new_change_ratio (re-arrangements at 0.5)"
mean = df.groupby(['label'])[column_name].mean()
std = df.groupby(['label'])[column_name].std()

mean.plot(kind='bar', yerr=std)
```
