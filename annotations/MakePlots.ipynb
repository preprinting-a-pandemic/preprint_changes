```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("../all_pairs - old.tsv", sep='\t')
df = df[df['Contains a 2 (binary)'].notna()]
df = df.reset_index(drop=True)
print (len(df))
```

```python
df.keys()
```

```python
import matplotlib.pyplot as plt

approaches = ['difflib standard change_ratio', 'Word_change_ratio',"change_outcomes"]
labels = ["change_outcomes",'Contains a 2 (binary)', 'Either 0, 1 or 2','Contains a 2']


for approach in approaches:
    fig, axs = plt.subplots(1,len(labels),figsize=(20,3))
    print (approach)

    for l in range(len(labels)):
        label = labels[l]
        df["label"] = df.apply(lambda row: str(row[label]), axis=1)
        df.head()
        mean = df.groupby(['label'])[approach].mean()
        std = df.groupby(['label'])[approach].std()

        mean.plot(kind='bar', yerr=std,ax=axs[l],title=label)
    
```
